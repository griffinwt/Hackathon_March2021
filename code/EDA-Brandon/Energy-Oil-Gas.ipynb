```python
def charRepeat(M = 0, char = '*'):
    return char*M

def FaresBoyfriendsFunction(F=0, Q='*'):
    minimum = 1
    count = 0
    a = charRepeat(minimum)
    #a = charRepeat
    for line in range(1, 2*F):
        if(line == 1) | (line ==2):
            d = charRepeat(line)
        elif line <= F:
            b = charRepeat(' ', line-2)
            d = a+b+a
        elif line < (2*F-1):
            count += 1
            b = charRepeat(' ', count)
            c = charRepeat(' ', F-(2+count))
            d = b + a+ c + a
        elif line == 2*F-1:
            count += 1
            b = charRepeat(' ', count)
            d = b + a
        print(d)
```
