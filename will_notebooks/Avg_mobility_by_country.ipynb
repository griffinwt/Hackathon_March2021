```python
#function to properly sort dfs and join them
def process_mobility_df(df):
    pd.to_datetime(df['Date']) #convert date from object
    df.sort_values(by='Date', inplace=True) #sort df by date
    new_df = pd.DataFrame(df.groupby('Date')['Value'].mean()).copy() #create new df sorted by date
    #new_df.set_index(new_df['Date'], inplace=True)
    return new_df
```

```python
mobility = process_mobility_df(ger_walk) #test
```

```python
mobility.rename(columns={'Value':'Value_ger_walk'}, inplace=True)
```

```python
mobility
```

```python
df_list = [ger_drive, ger_trans, #does not include ger_walk because that is the base of the "mobility" dataframe that we will be adding to
          ind_walk, ind_drive,
          rus_walk, rus_drive,
          us_walk, us_drive, us_trans,
          jap_walk, jap_drive, jap_trans] #list of all dfs to manipulate and join
```

```python
column_names = ['ger_drive', 'ger_trans', #list of column names to use in the new mobility df
               'ind_walk', 'ind_drive',
               'rus_walk', 'rus_drive',
               'us_walk', 'us_drive', 'us_trans',
               'jap_walk', 'jap_drive', 'jap_trans']
```

```python
for n in range(len(df_list)):
    mobility = mobility.join(process_mobility_df(df_list[n]), on = 'Date', how='outer', lsuffix='', rsuffix=f'_{column_names[n]}')
    #mobility = pd.merge(left=mobility, right=process_mobility_df(df), how='outer', suffixes=('', f'_{df}'))
   # mobility.merge(process_mobility_df(df), on = 'index', how = 'outer', suffixes=f'_{df}')
```
